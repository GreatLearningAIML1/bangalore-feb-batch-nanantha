```python
def Goal_diff_count (a,b):
    diff=a-b
    return (diff)

    
```

```python
laliga["Goaldifference"] = laliga.apply (lambda row:Goal_diff_count(row["GoalsFor"],row ["GoalsAgainst"]),axis=1)
print (laliga.reset_index()[["Team","Goaldifference"]])
```

```python
print ("Max", laliga[["Team","Goaldifference"]][laliga.Goaldifference == laliga.Goaldifference.max()])
```

```python
print ("Min", laliga[["Team","Goaldifference"]][laliga.Goaldifference == laliga.Goaldifference.min()])
```
