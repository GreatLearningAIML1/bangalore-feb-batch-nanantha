# mean of old scheme and new scheme column

```python
import pandas as pd
import numpy as np
```

```python
## Reading a dataset from a csv file using pandas.
data = pd.read_csv('data_14042019_project.csv')
data
```

```python
# mean of old scheme
d1=data.loc[:,"Old Scheme (in thousands)"].mean()
d1
#variance
d1v=data.loc[:,"Old Scheme (in thousands)"].var()
```

```python
# mean of new scheme
d2=data.loc[:,"New Scheme (in thousands)"].mean()
d2
#variance
d2v=data.loc[:,"New Scheme (in thousands)"].var()
```

```python
#five percent significance test
from scipy.stats import ttest_ind
```

```python
ttest_ind(data.loc[:,"Old Scheme (in thousands)"], data.loc[:,"New Scheme (in thousands)"])
```

```python
#p value > 0.05 , means p value has significantly increased output. accept null hypothesis that sales has increased
# the conclusion based on p value means that the new scheme has significantly improved output. sales has gone up with new scheme. 
# since its expensive, let us now focus on bottom-line
```

```python
data.insert(3, "expected sales",0) 
```

```python
# adding 5000 to old scheme for arriving at expected sales
data['expected sales']=data['Old Scheme (in thousands)']+5
```

```python
# ttest for new scheme and expected sales
ttest_ind(data.loc[:,"Old Scheme (in thousands)"], data.loc[:,"expected sales"])
```

```python
# the probability that the null hypothesis is wrong will be 66% - type 1 error for the data with average of 5000 pounds moe
# type 1 error between old and new scheme is 50%
```

```python
# power of the test based on old and new scheme without considering 5000 pounds increase in average hypothesis
# Calculating power of test for the energ dataset
import math 
sqrt=math.sqrt((((30-1)*d1v+(30-1)*d2v)) / 30+30-2)
p=(d1 - d2)/ sqrt
p
```
